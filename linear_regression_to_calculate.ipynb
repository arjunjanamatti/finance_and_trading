```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
import os

os.chdir("C:/Users/Arjun Janamatti/Downloads/")

# df = pd.read_csv("^NSEBANK Nifty_Latest.csv")
df = pd.read_csv("BANKNIFTY20APRFUT.csv")
df = df[::-1]

length_of_df = len(df.columns)
for i in range(2,6,1):
# get the adjusted close price
    df_new = df.iloc[:,[length_of_df-i]]
    print(df_new.head())

    # below is the variable, which defines the number of days to be predicted in future
    forecast_out = 64

    # create another column (the target or dependent variable) shifted "n" units up
    df_new["Prediction"] = df_new.iloc[:,[0]].shift(-forecast_out)


    # create the independent data set (X) 
    # convert the dataframe to a numpy array
    X = np.array(df_new.drop(["Prediction"],1))
    X = X[:-forecast_out]

    #  Create the dependent data set (y)
    #  Convert the dataframe to a numpy array (All of the values including the NAN)
    y = df_new.values
    #  get all the y values, except for the last "n" rows
    y = y[:,1]
    y = y[:-forecast_out]

    #  split the data into 80% training and 20% testing
    x_train, x_test, y_trian, y_test = train_test_split(X,y,test_size = 0.25)

    # create and train our model
    #  we will use support vector machine(Regressor)
    svr_rbf = SVR(kernel="rbf", C = 1e3, gamma = 0.1)
    svr_rbf.fit(x_train, y_trian)

    # testing model: score returns the coeffecient of determination R**2 of the prediction
    #  the best possible score is 1
    svm_confidence = svr_rbf.score(x_test, y_test)
    print("SVM confidence is :", svm_confidence)
    print()

    #  create and train the Linear regression model
    lr = LinearRegression()
    #  train the model
    lr.fit(x_train,y_trian)

    # testing model: score returns the coeffecient of determination R**2 of the prediction
    #  the best possible score is 1
    lr_confidence = lr.score(x_test, y_test)
    print("Linear Regression confidence is: ", lr_confidence)
    print()

    # set x_forecast equal to the last 30 rows of the original data set from the Adj. close
    x_forecast = np.array(df_new.drop(["Prediction"],1))[-forecast_out:]

    #  print the linear regression model predictions for the next n days
    lr_prediction = lr.predict(x_forecast)
    print("Linear regression prediction values are: ", lr_prediction)
    print()

    #  print the support vector regressor predictions for the next n days
    svm_prediction = svr_rbf.predict(x_forecast)
    print("SVR prediction values are: ", svm_prediction)

    np.savetxt("lr_prediction_{0}.txt".format(i), lr_prediction, fmt="%7.7f")
    np.savetxt("svm_prediction_{0}.txt".format(i), svm_prediction, fmt="%7.7f")
```
